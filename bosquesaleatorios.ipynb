```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
petalos = sns.load_dataset('iris')
```

```python
petalos
```

```python
lb = LabelEncoder()
```

```python
datos = lb.fit_transform(petalos["species"])
```

```python
datos 
```

```python
X = petalos.drop("species", axis=1)
y= datos
```

```python
X_entrena,X_prueba,y_entrena,y_prueba=train_test_split(X,y,test_size=0.8,random_state=46)
```

```python
modelos=[("regresion logistica",LogisticRegression()),
         ("arboles de decision",DecisionTreeClassifier()),("bosque aleatorio",RandomForestClassifier())]
```

```python
for nombre,modelo in modelos:
    modelo.fit(X_entrena,y_entrena)
    puntaje = modelo.score(X_prueba,y_prueba)
    print(f"{nombre}, {puntaje:.4f}")
```

```python
modelo_optimo= (RandomForestClassifier())
```

```python
modelo_optimo.fit(X_entrena,y_entrena)
```

```python
modelo_optimo.score(X_prueba,y_prueba)
```

```python
registro_prueba =pd.DataFrame ({"sepal_length":[2.5],"sepal_width":[2.5],"petal_length":[2.5],"petal_width":[1.5]},index=[0])
```

```python
prueba =modelo_optimo.predict_proba(registro_prueba)
```
